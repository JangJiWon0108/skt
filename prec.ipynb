```python
# 데이터 로딩

import os
import json

path_law="data/PREC/"
json_iist_law=[]

try:
    all_files=sorted(os.listdir(path_law))
    json_files=[file for file in all_files if file.endswith(".json")]
    
    for json_file in json_files:
        with open(path_law+"/"+json_file, "r", encoding="utf-8") as f:
            try:
                data=json.load(f)
                json_iist_law.append([json_file, data])

            except Exception as error:
                print("Json 읽는 중 Error")
                print(error)
    
    print("{} ---- [개수 : {}]".format(path_law, len(json_iist_law)))
        
except Exception as error:
    print("폴더 읽는 중 Error")
    print(error)
```

```python
# 케이스 확인

import pandas as pd

test_key="combined_content_dict_list"

test_result=[]

for file in json_iist_law:
    test_result.append(" | ".join(file[1][test_key][0].keys()))


df=pd.DataFrame(test_result, columns=["test"])

df["test"].value_counts()
```

```python
# key 값이 모두 있는지 확인

col_list=["prec_sentence", "prec_sentencing_date", "prec_judgment_summary", "prec_court_name", "prec_refer_precedent", "prec_refer_provision"]
col_content_list=[["plaintiff_appellant", "원고,피상고인"], ["defendant_appellee", "피고,상고인"], "judgment_order", "judgment_reason"]

for file in json_iist_law:
    for col in col_list:
        if col not in file[1]:
            print("{} - 존재하지 않음".format(col))
    try:
        content=file[1]["combined_content_dict_list"][0]
        for index in range(len(col_content_list)):

            if index==0:
                count=0
                if col_content_list[index][0] not in content:
                    count+=1
                    

                if col_content_list[index][1] not in content:
                    count+=1
                    print(col_content_list[index][0] not in content)
                    print(col_content_list[index][0])
                    print(content)
                    print(count)
                    
                if count==2:
                    print("{} - 존재하지 않음".format("원고"))

            elif index==1:
                count=0
                if col_content_list[index][0] not in content:
                    count+=1

                if col_content_list[index][1] not in content:
                    count+=1

                if count==2:
                    print("{} - 존재하지 않음".format("피고"))
            else:
                if col_content_list[index] not in content:
                    print("{} - 존재하지 않음".format(col_content_list[index]))

    except Exception as error:
        print(error)
```
